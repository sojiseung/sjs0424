```python
import pandas as pd
import numpy as np
st1 = pd.Series({'국어':100,'영어':80,'수학':90})
st2 = pd.Series({'국어':80,'영어':90,'수학':80})

print(st1, '\n')
print(st2)
```

```python
pd.DataFrame(st1+st2)
```

```python
# 두 학생의 과목별 점수로 사칙연산 수행
add = st1 + st2
sub = st1 - st2
mul = st1 * st2
div = round((st1 / st2),2)
df = pd.concat([add,sub,mul,div],axis=1)
print(df)
```

```python
df = pd.concat([add,sub,mul,div],axis=1)
df.columns = ['덧셈','뺼셈','곱셈','나눈셈']
# df.index.name = '과목'
df
```

```python
result = pd.DataFrame([add,sub,mul,div])
result
```

```python
result = pd.DataFrame([add,sub,mul,div],index=['덧셈','뺼셈','곱셈','나눈셈'])
result
```

```python
st1 = pd.Series({'국어':np.nan,'영어':80,'수학':90})
st2 = pd.Series({'수학':80,'국어':90})


add = st1 + st2
sub = st1 - st2
mul = st1 * st2
div = round((st1 / st2),2)

result = pd.DataFrame([add,sub,mul,div],
                     index=['덧셈','뺄셈','곱셈','나눗셈'])
result
```

Task3_0528. 주어진 DataFrame의 여러 열에 대해 각기 다른 Series를 더하고, 결과를 새로운 DataFrame으로 반환한 후 각 행의 합계를 계산하여 새로운 열에 추가하여 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [1, 2, 3, 4, 5],
    "B": [10, 20, 30, 40, 50],
    "C": [100, 200, 300, 400, 500]
})

# Series 생성
s1 = pd.Series([5, 5, 5, 5, 5])
s2 = pd.Series([10, 10, 10, 10, 10])
s3 = pd.Series([15, 15, 15, 15, 15])
df
```

```python
new_df = pd.DataFrame(
    {'A':(df['A']+s1),'B':(df['B']+s2),'C':(df['C']+s3)}
)
new_df['합계'] = new_df['A']+new_df['B']+new_df['C']
new_df
```

#### 그룹 연산
* 복잡한 데이터를 어떤 기준에 따라 여러 그룹으로 나누어서 관찰할 수 있으며 이런 방식으로 분할 처리하는 것을 그룹 연산이라 함

* 그룹 객체 만들기(분할)
* 그룹 연산 메소드(적용-결합)
